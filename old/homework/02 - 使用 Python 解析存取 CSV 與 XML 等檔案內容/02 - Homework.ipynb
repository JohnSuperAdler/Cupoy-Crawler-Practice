## 比較一下範例檔案中的「File I/O」與「CSV Reader」讀出來的內容有什麼差異

根據範例檔案的結果：

1. 取出班次一的每一個時間，印出來就好

2. 將班次一的每一個時間用一個變數保存

3. 將所有班次和其每一個時間用一個變數保存

(Hint： 2&3 要想一下用什麼的資料型態做整理比較適合)

```python
##### Module Import

import os
import sys
import csv
from pprint import pprint
#from urllib.request import urlretrieve
#import chardet
```

```python
path_csv = 'database/bus_schedule_example.csv'
```

```python
### File I/O 
### 1. 

print('File I/O')
print()

sche_open = open(path_csv, 'r', encoding='utf-8')
sche_txt = sche_open.read()
print(sche_txt)
print()

txt_by_line_li = sche_txt.split('\n')[:-1]
departure_1_idx = txt_by_line_li[0].split(',').index('班次1')
departure_1_time_li = []
for i in range(1, len(txt_by_line_li)):
    departure_1_time_li.append(txt_by_line_li[i].split(',')[departure_1_idx])

print(departure_1_time_li)

#sche_open.close()
```

```python
### File I/O 
### 2. 

print(f'Variable departure_1_time_li:\n{departure_1_time_li}')
```

```python
### File I/O 
### 3.

departure_all_idx = [x for x, y in enumerate(txt_by_line_li[0].split(',')) if '班次' in y]
num_bus = 1
departure_all_time_di = {}

for idx in departure_all_idx:
    temp_time_li = []
    for i in range(1, len(txt_by_line_li)):
        temp_time_li.append(txt_by_line_li[i].split(',')[idx])
    departure_all_time_di[num_bus] = temp_time_li
    num_bus +=1
    
pprint(f'Variable departure_all_time_di:\n{departure_all_time_di}')
```

```python
### csv Reader
### 1. 

print('csv Reader')
print()

sche_open = open(path_csv, 'r', encoding='utf-8', newline='')
rows = csv.reader(sche_open)
line_count = 0
departure_1_time_li = []

for r in rows:
    if line_count == 0:
        departure_1_idx = r.index('班次1')
    else:
        departure_1_time_li.append(r[departure_1_idx])
    line_count += 1
print(departure_1_time_li)

sche_open.close()
```

```python
### csv Reader
### 2. 

print(f'Variable departure_1_time_li:\n{departure_1_time_li}')
```

```python
### File I/O 
### 3.

line_count = 0
sche_open = open(path_csv, 'r', encoding='utf-8', newline='')
rows = csv.reader(sche_open)
for r in rows:
    #print('into rows')
    if line_count == 0:
        departure_all_idx = [x for x, y in enumerate(r) if '班次' in y]
    else:
        break
    line_count += 1
#print(departure_all_idx)
sche_open.close()

line_count = 0
num_bus = 1
departure_all_time_di = {}

for idx in departure_all_idx:
    temp_time_li = []
    sche_open = open(path_csv, 'r', encoding='utf-8', newline='')
    rows = csv.reader(sche_open)
    for r in rows:
        #print('into rows')
        if line_count != 0:
            #print('Not Zero')
            temp_time_li.append(r[idx])
        else:
            #print('Zero')
            pass
        line_count += 1
    departure_all_time_di[num_bus] = temp_time_li
    num_bus +=1
    sche_open.close()
        
sche_open.close()
    
pprint(f'Variable departure_all_time_di:\n{departure_all_time_di}')
```

## 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異

根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？

2. 請取出每一個地區所記錄的第一個時間點跟溫度

3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
path_csv_2 = 'database/64_72hr_CH.xml'
```

```python
### File I/O 
### 1. 

print('File I/O')

report_open = open(path_csv_2, 'r', encoding='utf-8')
report_txt = report_open.read()
report_open.close()
#print(report_txt)
#print()

by_loc_li = report_txt.split('<location>')
by_loc_li = [x for x in by_loc_li if '<description>溫度</description>' in x]
print(f'Number of locations with temperature data: {len(by_loc_li)}')
```

```python
### File I/O 
### 2.

for i in range(len(by_loc_li)):
    # Get location name
    loc_name = by_loc_li[i].split('<locationName>')[1].split('</locationName>')[0]    
    by_we_li = by_loc_li[i].split('<weatherElement>')
    by_we_li = [x for x in by_we_li if '<description>溫度</description>' in x]
    time = by_we_li[0].split('<dataTime>')[1].split('</dataTime>')[0]
    temp = by_we_li[0].split('<value>')[1].split('</value>')[0]
    print(f'{loc_name}: {time}, {temp} degC')
```

```python
### File I/O 
### 3.

first_loc = by_loc_li[0]
loc_name = by_loc_li[0].split('<locationName>')[1].split('</locationName>')[0]  
print(loc_name)
by_temp_we = [x for x in first_loc.split('<weatherElement>') if '<description>溫度</description>' in x][0]
by_time_li = [x for x in by_temp_we.split('<time>') if '<value>' in x]
for i in range(len(by_time_li)):
    time = by_time_li[i].split('<dataTime>')[1].split('</dataTime>')[0]
    temp = by_time_li[i].split('<value>')[1].split('</value>')[0]
    print(f'{i}: {time}, {temp} degC')
```

```python
import xmltodict
```

```python
### xmltodict
### 1.

print('xmltodict')
print()

report_open = open(path_csv_2, 'r', encoding='utf-8')
data_di = dict(xmltodict.parse(report_open.read()))
report_open.close()
print(f'type(data_di): {type(data_di)}')

by_loc_li = data_di['cwbopendata']['dataset']['locations']['location']
print(len(by_loc_li))
print(type(by_loc_li))
temp_data_count = 0
for i in range(len(by_loc_li)):
    if by_loc_li[i]['weatherElement'][0]['description'] == '溫度':
        temp_data_count +=1
        
print(f'Number of locations with temperature data: {temp_data_count}')
```

```python
### xmltodict
### 2.

for i in range(len(by_loc_li)):
    loc_name = by_loc_li[i]['locationName']
    print(loc_name)
    by_time_li = by_loc_li[0]['weatherElement'][0]['time']#[0]['dataTime']
    for j in range(len(by_time_li)):
        if j != 0:
            continue
        time = by_time_li[j]['dataTime']
        temp = by_time_li[j]['elementValue']['value']
        print(f' - {time}, {temp} degC')
```
