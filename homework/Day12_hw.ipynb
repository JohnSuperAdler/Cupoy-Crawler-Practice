1. https://ic.tpex.org.tw/index.php
2. https://www.zhihu.com/explore

- 利用開發者工具找出這兩個網站的 HTTP 網址，觀察其所帶的 Headers 有哪些
- 測試有無帶 Headers 發送 request 的差別


```python
import requests
```

### (1) https://ic.tpex.org.tw/index.php

- 不帶 headers


```python
url = 'https://ic.tpex.org.tw/index.php'

r1 = requests.get(url)
r1.encoding = 'utf-8'
print(r1.text)
```

- 帶 headers


```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'
}
r2 = requests.get(url, headers=headers)
r2.encoding = 'utf-8'
print(r2.text)
```

### (2) https://www.zhihu.com/explore


- 不帶 headers

```python
url = 'https://www.zhihu.com/explore'
r = requests.get(url)
r.encoding = 'utf-8'
print(r.text)
print(r.status_code)
```

- 帶 headers

```python
url = 'https://www.zhihu.com/explore'
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'
}
r = requests.get(url, headers=headers)
r.encoding = 'utf-8'
print(r.text)
```

```python
print(r.status_code)
```

zhihu 網頁需要使用 user-agent 的 fake header 才能收到正常的網頁html碼。
